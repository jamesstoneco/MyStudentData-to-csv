```python
cols = ['Loan Type:',
'Loan Award ID:',
'Loan Attending School Name:',
'Loan Date:',
'Loan Amount:',
'Loan Disbursed Amount:',
'Loan Outstanding Principal Balance:',
'Loan Outstanding Interest Balance:',
'Loan Interest Rate Type:',
'Loan Interest Rate:',
'Loan Contact Name:',
'dummy', # HACK: because I am getting Loan Contact Name coming in 3 times on some entries
'dummy']
```

```python
delimiter = 'Loan Type:'
loans = open('MyStudentData.txt', 'r').read().split(delimiter)
```

```python
loans.pop(0) # remove header info
```

```python
proc_loans = []
for l in loans:
    proc_loans.append((delimiter + l).split('\n'))
```

```python
data = []
for p in proc_loans:
    temp_loan = []
    for c in cols:
        for line in p:
            if line.startswith(c):
                    temp_loan.append(line.split(':')[1])
    data.append(temp_loan)

data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_records(data, columns=cols, coerce_float=True)
```

```python
df.convert_dtypes(infer_objects=True, convert_string=True, convert_integer=True, convert_boolean=True)
```
